Merge with:
- nlp listing file (topics 0-7)
- nlp review file

```python
# nlp listings
nlp_listings_0_7 = pd.read_csv("/Users/satanhaha/Documents/MS ADS/ML 1/ML1 Final Project/Airbnb-Renting-Optimizer/data_processed/listings_with_nlp_features.csv")
nlp_listings_0_7.drop(columns=['description','tokens_final'], inplace=True)
nlp_listings_0_7.rename(columns={'id': 'listing_id'}, inplace=True)

# nlp reviews
nlp_reviews = pd.read_csv("/Users/satanhaha/Documents/MS ADS/ML 1/ML1 Final Project/Airbnb-Renting-Optimizer/data_processed/Review Topics Final.csv")
```

```python
master_v5 = master_v4.merge(
    nlp_listings_0_7,
    on="listing_id",
    how="left",
    validate="one_to_one"
)

master_v5 = master_v5.merge(
    nlp_reviews,
    on="listing_id",
    how="left",
    validate="one_to_one"
)

master_v5.head()
```
