```python
import pandas as pd
last_df = pd.read_csv("Chicago_Airbnb_Master_v2.csv")

last_df.head()
```

```python
geo_df = pd.read_csv("Geospatial_Engineering_v3.csv")

geo_df.head()
```

```python
if "id" in geo_df.columns:
    geo_df = geo_df.rename(columns={"id": "listing_id"})

geo_df.head()
```

```python
print("master_v2 columns:")
print(last_df.columns.tolist())

print("\ngeo_master columns:")
print(geo_df.columns.tolist())
```

```python
geo_df = geo_df.drop("price", axis=1)

geo_df.head()
```

```python
master_v3 = last_df.merge(
    geo_df,
    on="listing_id",
    how="left",
    validate="one_to_one"
)

print(master_v3.columns.tolist())
```

```python
print("Rows:", master_v3.shape[0])
print("Duplicate listing_id:", master_v3["listing_id"].duplicated().sum())
print("Missing cluster labels:", master_v3["dist_navy_pier_miles"].isna().sum())
```

```python
master_v3.to_csv("Chicago_Airbnb_Master_v3.csv", index=False)
```
