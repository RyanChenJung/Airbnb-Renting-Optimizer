```python
gdf_joined = gpd.sjoin(
    gdf_listings,
    geo[["neighbourhood", "geometry"]],
    how="left",
    predicate="within"
)
```

```python
gdf_joined[["listing_id", "Cluster_ID", "neighbourhood"]].head()
```

### Local Price Power

How expensive is this neighborhood relative to the city?

```python
neigh_price = (
    gdf_joined
    .groupby("neighbourhood")["price"]
    .median()
    .reset_index()
    .rename(columns={"price": "neigh_median_price"})
)
```

```python
city_median_price = gdf_joined["price"].median()
```

```python
neigh_price["neigh_price_index"] = (
    neigh_price["neigh_median_price"] / city_median_price
)

neigh_price.head()
```

### Local Demand Strength (Occupancy Proxy)

Captures neighborhood demand

```python
neigh_occ = (
    gdf_joined
    .groupby("neighbourhood")["occupancy_proxy"]
    .mean()
    .reset_index()
    .rename(columns={"occupancy_proxy": "neigh_avg_occ"})
)
```

### Competitive Intensity

Captures supply pressure

```python
neigh_comp = (
    gdf_joined
    .groupby("neighbourhood")["listing_id"]
    .count()
    .reset_index()
    .rename(columns={"listing_id": "neigh_listing_count"})
)

import numpy as np
neigh_comp["neigh_log_competition"] = np.log1p(
    neigh_comp["neigh_listing_count"]
)
```

## Relative Positioning (High Signal)

Captures listing pricing strategy

```python
master_geo = gdf_joined.merge(neigh_price, on="neighbourhood", how="left")

master_geo["price_vs_neigh_median"] = (
    master_geo["price"] / master_geo["neigh_median_price"]
)
```

## Distance to CBD (Economic Gradient)

```python
downtown_lat = 41.8781
downtown_lon = -87.6298
```

```python
!pip install geopy
```

```python
from geopy.distance import geodesic

master_geo["distance_to_downtown_km"] = master_geo.apply(
    lambda row: geodesic(
        (row["latitude"], row["longitude"]),
        (downtown_lat, downtown_lon)
    ).km,
    axis=1
)
```

```python
master_geo.columns.tolist()
```

```python
# Merge occupancy
master_geo = master_geo.merge(
    neigh_occ,
    on="neighbourhood",
    how="left",
    validate="many_to_one"
)

# Merge competition
master_geo = master_geo.merge(
    neigh_comp[["neighbourhood", "neigh_log_competition"]],
    on="neighbourhood",
    how="left",
    validate="many_to_one"
)
```
